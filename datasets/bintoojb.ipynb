```python
import numpy as np

def read_bin_file(file_path):
    """Reads a binary file and interprets it as 3D coordinates."""
    with open(file_path, 'rb') as file:
        data = np.frombuffer(file.read(), dtype=np.float32).reshape(-1, 4)[:, :3]
    return data

def write_point_cloud_to_obj(data, output_path):
    """Writes point cloud data to an OBJ file."""
    with open(output_path, 'w') as file:
        for point in data:
            file.write(f"v {point[0]} {point[1]} {point[2]}\n")

def process_bin_to_obj(bin_file_path, obj_output_path):
    """Converts a .bin file of 3D coordinates to an OBJ file."""
    point_cloud_data = read_bin_file(bin_file_path)
    write_point_cloud_to_obj(point_cloud_data, obj_output_path)
    print(f"Data from {bin_file_path} has been processed and saved to {obj_output_path}")

# Paths to your .bin file and output .obj file
bin_file_path = '/Users/lijiarong/Downloads/0000000000.bin'
obj_output_path = '/Users/lijiarong/Downloads/output.obj'


# Call the function to process the file
process_bin_to_obj(bin_file_path, obj_output_path)
```
